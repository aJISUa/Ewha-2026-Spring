```python
t1='''설영은
신수빈
이다영
정혜교
김나현
박나림
조채윤
조상은
신은서
신성현
최윤서
이나겸
손하늘
최준희'''
teams=[n for n in t1.split('\n')]
print(teams)
```

```python
import random

go = random.sample(teams, 10)
print(go)
print(set(teams)-set(go))
```
